## 1-1. ë¦¬ìŠ¤íŠ¸ë¥¼ ì§ì ‘ ëŒê¸° vs ì¸ë±ìŠ¤ë¡œ ëŒê¸°
- `range(len(data))`ì™€ `for x in data`

```python
data = ["A", "B", "C", "D"]

# ğŸ‘‰ğŸ» ë°©ë²•1: ìš”ì†Œ ì¤‘ì‹¬ (ê°€ì¥ íŒŒì´ì¬ ë‹µê³  ì½ê¸° í¸í•¨)
# "ë°ì´í„°ë¥¼ í•˜ë‚˜ì”© êº¼ë‚´ ì“°ê² ë‹¤"
for char in data:
  print(f"êº¼ë‚¸ ë¬¸ì: {char}")

# ğŸ‘‰ğŸ» ë°©ë²•2: ìœ„ì¹˜ ì¤‘ì‹¬ (ì¸ë±ìŠ¤ í™œìš©)
# "ëª‡ ë²ˆì§¸ì¸ì§€ê°€ ì¤‘ìš”í•˜ê³ , ì›ë³¸ ë¦¬ìŠ¤íŠ¸ì˜ ê°’ì„ ë°”ê¿”ì•¼í•  ë•Œ"
for i in range(len(data)):
  print(f"{i}ë²ˆì§¸ ìœ„ì¹˜ì˜ ë¬¸ìëŠ” {data[i]} ì…ë‹ˆë‹¤.")
```

## 1-2. ë‘ ë¦¬ìŠ¤íŠ¸ë¥¼ ë™ì‹œì— ëŒë©°(zip) ìˆœì„œ(enumerate) ë§¤ê¸°ê¸°
- ì‹¤ì „ì—ì„œëŠ” ì—¬ëŸ¬ ë¦¬ìŠ¤íŠ¸ë¥¼ ë™ì‹œì— ëŒê±°ë‚˜ ì¸ë±ìŠ¤ê°€ í•„ìš”í•œ ê²½ìš°ê°€ ë§ë‹¤

```python
# ğŸ§© ë‘ ë¦¬ìŠ¤íŠ¸ë¥¼ ë™ì‹œì— ëŒë©°(zip) ìˆœì„œ(enumerate)ë§¤ê¸°ê¸°
names = ["Alice", "Bob", "Charlie"]
scores = [85, 90, 95]

for i, (name, score) in enumerate(zip(names, scores), start=1):
  print(f"{i}ë²ˆ í›„ë³´: {name}ë‹˜ì˜ ì ìˆ˜ëŠ” {score}ì ì…ë‹ˆë‹¤.")
```

## 2-1. Nested if(ì¤‘ì²© ì¡°ê±´)ì˜ ìœ„í—˜ì„±

```python
# [ë¬¸ì œ ìƒí™©] 1~100 ìˆ«ì ì¤‘, 3ì˜ ë°°ìˆ˜ì´ë©´ì„œ ì§ìˆ˜ì¸ ê²½ìš°ë§Œ ì¶œë ¥ (ë‹¨, 50 ì´ìƒì€ ì¤‘ë‹¨)
# ì•ˆ ì¢‹ì€ ì˜ˆ (Indent ì§€ì˜¥)
for n in range(1, 101):
    if n < 50:
        if n % 3 == 0:
            if n % 2 == 0:
                print(f"ì°¾ì•˜ë‹¤! {n}")
    else:
        break
```

## 2-2. ê°€ë“œì ˆë¡œ ë¦¬íŒ©í† ë§
- ë‚´ê°€ ì›í•˜ì§€ ì•ŠëŠ” ì¡°ê±´ì„ ë¨¼ì € ê²€ì‚¬í•˜ì—¬ ì§€ìš°ê¸°

```python
# ê°€ë“œì ˆ í™œìš© (ì¢‹ì€ ì˜ˆ)
for n in range(1, 101):
  if n >= 50:  # 1ìˆœìœ„ ê°€ë“œ: ì¤‘ë‹¨ ì¡°ê±´
    break
  if n % 3 != 0:  # 2ìˆœìœ„ ê°€ë“œ: 3ì˜ ë°°ìˆ˜ê°€ ì•„ë‹ˆë©´ íŒ¨ìŠ¤
    continue
  if n % 2 != 0:  # 3ìˆœìœ„ ê°€ë“œ: í™€ìˆ˜ë©´ íŒ¨ìŠ¤
    continue
  # ëª¨ë“  ê²€ì‚¬ë¥¼ í†µê³¼í•œ ìˆ«ìë§Œ
  print(f"ì°¾ì•˜ë‹¤! {n}")
```

# 3. ì¡°ê±´ì— ë§ëŠ” ë°ì´í„° ì¶”ì¶œí•˜ê¸°
- ë°ì´í„° ë¶„ì„ì— ê°€ì¥ ë§ì´ ì‚¬ìš©ë˜ëŠ” íŒ¨í„´ì€ **ì¡°ê±´ í•„í„°ë§ í›„ ìƒˆë¡œìš´ ë¦¬ìŠ¤íŠ¸ ë§Œë“¤ê¸°**

## íŒ¨í„´ A: ì „í†µì  ë°©ì‹

```python
raw_data = [1, 5, 20, 45, 12, 60, 8, 100]
filtered_data = []

for val in raw_data:
  if val < 10:  # 10ë³´ë‹¤ ì‘ìœ¼ë©´ ë¬´ì‹œ
    continue
  if val > 80:  # 80ë³´ë‹¤ í¬ë©´ ë£¨í”„ ì•„ì˜ˆ ì¢…ë£Œ
    break
  filtered_data.append(val)
print(f"ìµœì¢… í•„í„°ë§ ê²°ê³¼: {filtered_data}")
```

## íŒ¨í„´ B: ë¦¬ìŠ¤íŠ¸ ì»´í”„ë¦¬í—¨ì…˜

```python
filtered_data = [val for val in raw_data if 10 <= val <= 80]
print(f"ìµœì¢… í•„í„°ë§ ê²°ê³¼: {filtered_data}")
```

ì—°ìŠµ 1. ë³€í˜•í•´ì„œ ë‹´ê¸°(Mapping)

```python
words = ["apple", "banana", "cherry"]

# ëª¨ë“  ë‹¨ì–´ ëŒ€ë¬¸ìë¡œ ë§Œë“¤ê¸°
upper_words = [w.upper() for w in words]
print(upper_words)
```

ì—°ìŠµ 2. if-elseë¥¼ í¬í•¨í•œ ì¡°ê±´ë¶€ ë‹´ê¸°
- scores = [45, 88, 72, 50]

```python
# íŠ¹ì • ì¡°ê±´ì— í•´ë‹¹í•˜ë©´ Aë¥¼ ë‹´ê³  ì•„ë‹ˆë©´ Bë¥¼ ë‹´ì•„ë¼
scores = [45, 88, 72, 50]

# 60ì  ì´ìƒì´ë©´ 'Pass', ì•„ë‹ˆë©´ 'Fail'
result = ["Pass" if s >= 60 else "Fail" for s in scores]
print(result)
```

```python
names = ["Alice", "Bob"]
scores = [85, 90]

# (ì´ë¦„: ì ìˆ˜) í˜•íƒœì˜ ë¬¸ìì—´ ë¦¬ìŠ¤íŠ¸ ë§Œë“¤ê¸°
summary = [f"{n}: {s}" for n, s in zip(names, scores)]
print(summary)
```
