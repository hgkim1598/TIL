### 파이썬 리스트 내장함수
| 함수명 | 설명 | 예시 | 출력 |
|--------|------|------|------|
| min | 리스트 내 최소값 반환 | `min([-1, 0, 1, 2])` | `-1` |
| max | 리스트 내 최대값 반환 | `max(['ㄱ', 'ㄴ', 'ㄷ'])` | `'ㄷ'` |
| len | 리스트의 길이 반환 | `len(['환자1', '환자2'])` | `2` |
| sum | 리스트 내 모든 값을 더하기 (숫자형만) | `sum([1, 3, 4, 5, 9])` | `22` |
| all | 리스트의 모든 값이 참인지 확인 | `all([0, 1, 2, 6, 7])` | `False` |
| any | 리스트 내의 값이 하나라도 참인지 확인 | `any([0, 1, 0, 0, 0])` | `True` |
| enumerate | 리스트 내 인덱스와 값을 함께 반환 (반복문에서 활용) | `list(enumerate([100, 3]))` | `[(0, 100), (1, 3)]` |
| range | 특정 숫자 범위(시작, 끝)의 값들을 순차적으로 반환 (반복문에서 활용) | `list(range(1, 3))` | `[1, 2]` |
| sorted | 리스트의 순서를 정렬하기 | `sorted([2, 1, 0])` | `[0, 1, 2]` |



| 함수명 | 설명 |
|--------|------|
| append | 리스트 마지막에 값 추가 |
| sort | 리스트 내 값들의 순서 정렬하기 |
| reverse | 리스트 내 값들의 순서를 역으로 정렬하기 |
| count | 리스트 내 특정 값의 개수 세기 |
| index | 리스트 내 특정 값의 인덱스 불러오기 |
| pop | 리스트 내 특정 인덱스의 값 제거하기 |
| remove | 리스트 내 지정 값을 가진 요소 제거하기 |
| insert | 리스트 내 특정 위치에 값 삽입하기 |


```python
# 알파벳 순 오름차순
hospitals = ["서울대병원", "신촌세브란스병원", "서울성모병원", "아산병원", "고려대병원"]
hospitals.sort()
print(hospitals)

# 알파벳 순 내림차순
hospitals.sort(reverse=True)
print(hospitals)

# Sorted 활용 예시
sorted(hospitals)
print(hospitals)

hospitals = sorted(hospitals)
print(hospitals)
```

```python
# 리스트 순서 뒤집기
hospitals = ["서울대병원", "신촌세브란스병원", "서울성모병원", "아산병원", "고려대병원"]
hospitals.reverse()

print(hospitals)
```

```python
# 데이터의 리스트 내 index 찾기
medicines = ["A", "B", "E", "H", "S", "E"]
print(medicines.index("E"))
```

```python
# 리스트 내 데이터 삽입하기
hospitals = ["서울대병원","신촌세브란스병원","서울성모병원","아산병원","고려대병원"]
hospitals.insert(0, "차병원")
hospitals.insert(len(hospitals), "서북병원")
print(hospitals)
```

```python
# Q: 아래의 약물처방 리스트를 내림차순으로 정렬하고, A약물의 사용 횟수를 출력하시오
medicines = ["A","B","E","H","S","E","A","B","A","H","S","S","A","E","S","T","A","B","A"]

# sorted_medicines = medicines.sort(reverse=True) -> 값을 반환하지 않고 변환만 함
sorted_medicines = sorted(medicines, reverse=True)
count_A = medicines.count("A")

print(f"리스트 내림차순: {sorted_medicines}")
print(f"A의 개수: {count_A}")
```
