## 변수
| 구분 | 클래스 | 설명 | 예시 |
|------|--------|------|------|
| 숫자형 | int | 정수형 (소수점을 가지지 않은 숫자) | 1, -10, 0 |
| 숫자형 | float | 실수형 (소수점이 포함된 숫자) | 3.14, -2.3, 3.14E10, 3.14e-10 |
| 논리형 | bool | 부울형 (True 혹은 False) | True, False |
| 군집형 | str | 문자열 (문자, 단어 등으로 구성된 문자들의 집합 / 순서가 존재함) | `'a'`, `"환자1"`, `'''병원'''`, `"""hello world!"""` |
| 군집형 | list | 배열 (숫자와 문자열을 순서대로 저장한 데이터 구조 / 순서가 존재함) | `[1, 2, 3, 'one', ['two', 'three']]` |
| 군집형 | tuple | 읽기전용 배열 (숫자와 문자열을 저장한 데이터 구조 / 순서가 존재함 / 딕셔너리의 키로 사용 가능함 / 생성·삭제·수정 불가) | `(1, 2, 3, 'H', 'e', 'l', 'l', 'o')` |
| 군집형 | set | 집합 (집합의 원소들을 표현한 데이터 구조 / 순서 존재하지 않음 / 중복 값 허용하지 않음) | `{1, 2, 3, 'H', 'e', 'l', 'o'}` |
| 군집형 | dict | 사전형태 (이름(key)과 내용(value)의 대응관계를 나타낸 데이터 구조 / 순서 존재하지 않음 / Key는 고유한 값) | `{'gender': 'male', 'age': 25}` |
