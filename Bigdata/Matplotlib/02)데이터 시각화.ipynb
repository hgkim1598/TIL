# 여러 그래프

```python
# y1 = 2x^2
# y2 = 5x + 30
# y3 = 4x^2 + 10

import numpy as np
x = np.arange(-4.5, 5, 0.5)
y1 = 2*x**2
y2 = 5*x+30
y3 = 4*x**2+10
```

### 여러개의 함수 그리기

```python
plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
plt.show()
```

```python
plt.plot(x,y1)  # 처음 그리던 함수를 수행하면 그래프가 생성
plt.figure()  # 새로운 그래프 창을 생성
plt.plot(x, y2)  # 새로운 그래프 창에 그림

plt.show()
```
