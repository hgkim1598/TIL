# 위의 세 조건을 결합해보자

여러개의 복합 조건도 결합해 적용할 수 있다
1) and 조건일 때는 &

2) or 조건일 때는 |

3) Not 조건일 때는 ~

```python
# Q1. 60세 초과이며, 여성이고, 선실 등급 1등급에 자리했던 승객 구하기 => 2명
```

```python
titan_df[titan_df['Age'] > 60 & titan_df.Sex=='female' & titan_df.Pclass==1]
```

```python
# 강사님 정답
# 1. 멋있진 않지만 틀릴 일이 없다!, 가독성이 좋다
cond1 = titan_df.Age > 60
cond2 = titan_df.Sex == 'female'
cond3 = titan_df.Pclass == 1

titan_df[cond1 & cond2 & cond3]
```

```python
# 2. 한 줄에 쓰기
# 소괄호 활용!
titan_df[(titan_df['Age'] > 60) & (titan_df.Sex=='female') & (titan_df.Pclass==1)]
```

```python
# Q2. Q1번을 부정문으로 써보세요 => 889명
```

```python
# 1.의 부정 => 문제를 위한 문제일 뿐 이렇게 코딩하지 말자
cond1 = titan_df.Age > 60
cond2 = titan_df.Sex == 'female'
cond3 = titan_df.Pclass == 1

titan_df[~(cond1 & cond2 & cond3)]  # 이렇게 하지마
# and의 반대는 or?
```

```python
# 2.
titan_df[~((titan_df['Age'] > 60) & (titan_df.Sex=='female') & (titan_df.Pclass==1))]
```
